```python
import pickle
import matplotlib.pyplot as plt
```

```python
def plot_training_history_baseline(history_path, final_epoch=1000):
    with open(history_path, "rb") as f:
        hist = pickle.load(f)

    epochs = range(1, len(hist['loss'])+1)
    plt.figure(figsize=(10, 10))

    plt.subplot(2,1,1)
    plt.plot(epochs, [x*100 for x in hist['categorical_accuracy']], color='navy')
    plt.plot(epochs, [x*100 for x in hist['val_categorical_accuracy']], color='crimson')
    plt.ylabel('categorical accuracy\n(%)', fontsize=18, fontweight='bold', labelpad=20)
    plt.yticks(fontsize=13)
    plt.xticks(fontsize=13)
    plt.xlim(0, max(epochs))
    plt.ylim(0, 100)
    plt.text(600, 81, 'validation test', color='crimson', fontsize=17, fontweight='bold')
    plt.text(600, 75, 'training set', color='navy', fontsize=17, fontweight='bold')
    plt.axvline(x=final_epoch, color='gray', linestyle='--', linewidth=1.5, alpha=0.6)
    plt.text(final_epoch+23, 80, f'final model\n{final_epoch} epochs', color='gray', fontsize=14, fontweight='bold', ha='left')
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)

    plt.subplot(2,1,2)
    plt.plot(epochs, hist['loss'], color='navy')
    plt.plot(epochs, hist['val_loss'], color='crimson')
    plt.ylabel('loss', fontsize=18, fontweight='bold')
    plt.xlabel('epoch', fontsize=18, fontweight='bold')
    plt.yticks(fontsize=13)
    plt.xticks(fontsize=13)
    plt.xlim(0, max(epochs))
    plt.ylim(0, 2)
    plt.text(400, 0.82, 'validation test', color='crimson', fontsize=17, fontweight='bold')
    plt.text(400, 0.58, 'training set', color='navy', fontsize=17, fontweight='bold')
    plt.axvline(x=final_epoch, color='gray', linestyle='--', linewidth=1.5, alpha=0.6)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)

    plt.tight_layout()
    plt.show()

plot_training_history_baseline("trained_model_baseline/training_history_baseline.pkl", final_epoch=1000)
```

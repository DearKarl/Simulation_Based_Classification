```python
import pickle
import matplotlib.pyplot as plt
```

```python
def plot_training_history_baseline(history_path, final_epoch=1000):
    with open(history_path, "rb") as f:
        hist = pickle.load(f)

    epochs = range(1, len(hist['loss'])+1)
    final_train_acc = hist['categorical_accuracy'][-1] * 100
    final_val_acc = hist['val_categorical_accuracy'][-1] * 100
    final_train_loss = hist['loss'][-1]
    final_val_loss = hist['val_loss'][-1]

    plt.figure(figsize=(10, 10))

    plt.subplot(2, 1, 1)
    plt.plot(epochs, [x * 100 for x in hist['categorical_accuracy']], color='navy')
    plt.plot(epochs, [x * 100 for x in hist['val_categorical_accuracy']], color='crimson')
    plt.ylabel('categorical accuracy (%)', fontsize=14, fontweight='bold', labelpad=15)
    plt.yticks(fontsize=11)
    plt.xticks(fontsize=11)
    plt.xlim(0, 1000)
    plt.ylim(0, 100)
    plt.text(600, 82, f'validation test ({final_val_acc:.2f}%)', color='crimson', fontsize=13, fontweight='bold')
    plt.text(600, 75, f'training set ({final_train_acc:.2f}%)', color='navy', fontsize=13, fontweight='bold')
    plt.axvline(x=final_epoch, color='gray', linestyle='--', linewidth=1.2, alpha=0.6)
    plt.text(final_epoch+20, 80, f'final model\n{final_epoch} epochs', color='gray', fontsize=12, fontweight='bold', ha='left')
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)

    plt.subplot(2, 1, 2)
    plt.plot(epochs, hist['loss'], color='navy')
    plt.plot(epochs, hist['val_loss'], color='crimson')
    plt.ylabel('loss', fontsize=14, fontweight='bold', labelpad=15)
    plt.xlabel('epoch', fontsize=14, fontweight='bold')
    plt.yticks(fontsize=11)
    plt.xticks(fontsize=11)
    plt.xlim(0, 1000)
    plt.ylim(0, 3)
    plt.text(600, 0.8, f'validation test ({final_val_loss:.4f})', color='crimson', fontsize=13, fontweight='bold')
    plt.text(600, 0.6, f'training set ({final_train_loss:.4f})', color='navy', fontsize=13, fontweight='bold')
    plt.axvline(x=final_epoch, color='gray', linestyle='--', linewidth=1.2, alpha=0.6)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)

    plt.suptitle("Training and Validation Accuracy/Loss over Epochs (Baseline Model with Late-Average Fusion)",
                 fontsize=16, fontweight="bold", y=1.02)

    plt.tight_layout()
    plt.show()

plot_training_history_baseline("trained_model_baseline/training_history_baseline.pkl", final_epoch=1000)
```

```python
import pickle
import matplotlib.pyplot as plt

def plot_all_histories(histories, final_epochs, model_names, save_path=None):
    fig, axes = plt.subplots(2, 4, figsize=(32, 14))  # 24
    plt.subplots_adjust(hspace=0.25, wspace=0.25)

    #  Accuracy  & Loss 
    # [(val_x_frac, val_y), (train_x_frac, train_y)]
    acc_text_positions = {
        "Baseline Model with Late-Average Fusion": [(0.35, 85), (0.35, 77)],
        "Attention-Weighted Fusion Model":         [(0.35, 91), (0.35, 83)],
        "Gate-Weighted Additive Fusion Model":     [(0.35, 60), (0.35, 52)],
        "Hadamard Bilinear Fusion Model":          [(0.35, 60), (0.35, 52)],
    }
    loss_text_positions = {
        "Baseline Model with Late-Average Fusion": [(0.35, 0.85), (0.35, 0.60)],
        "Attention-Weighted Fusion Model":         [(0.35, 1.45), (0.35, 1.20)],
        "Gate-Weighted Additive Fusion Model":     [(0.35, 1.35), (0.35, 1.10)],
        "Hadamard Bilinear Fusion Model":          [(0.35, 1.35), (0.35, 1.10)],
    }

    for i, (history_path, final_epoch, model_name) in enumerate(zip(histories, final_epochs, model_names)):
        with open(history_path, "rb") as f:
            hist = pickle.load(f)

        epochs = range(1, len(hist['loss'])+1)
        final_train_acc = hist['categorical_accuracy'][-1] * 100
        final_val_acc = hist['val_categorical_accuracy'][-1] * 100
        final_train_loss = hist['loss'][-1]
        final_val_loss = hist['val_loss'][-1]

        # ===  Accuracy  ===
        ax1 = axes[0, i]
        ax1.plot(epochs, [x * 100 for x in hist['categorical_accuracy']], color='navy')
        ax1.plot(epochs, [x * 100 for x in hist['val_categorical_accuracy']], color='crimson')
        ax1.set_ylabel("categorical accuracy (%)", fontsize=25, fontweight='bold', labelpad=5)
        ax1.set_xlim(0, final_epoch)
        ax1.set_ylim(0, 100)
        ax1.tick_params(axis='both', labelsize=20)
        ax1.axvline(x=final_epoch, color='gray', linestyle='--', linewidth=1.2, alpha=0.6)

        #  
        (acc_val_xfrac, acc_val_y), (acc_tr_xfrac, acc_tr_y) = acc_text_positions.get(
            model_name, [(0.60, 82), (0.60, 75)]
        )
        ax1.text(acc_val_xfrac * final_epoch, acc_val_y, f'validation ({final_val_acc:.2f}%)',
                 color='crimson', fontsize=20, fontweight='bold')
        ax1.text(acc_tr_xfrac * final_epoch, acc_tr_y, f'training ({final_train_acc:.2f}%)',
                 color='navy', fontsize=20, fontweight='bold')

        ax1.spines['top'].set_visible(False)
        ax1.spines['right'].set_visible(False)
        ax1.set_title(model_name, fontsize=20, pad=25, fontweight='bold')

        # ===  Loss  ===
        ax2 = axes[1, i]
        ax2.plot(epochs, hist['loss'], color='navy')
        ax2.plot(epochs, hist['val_loss'], color='crimson')
        ax2.set_ylabel("loss", fontsize=25, fontweight='bold', labelpad=15)
        ax2.set_xlabel("epoch", fontsize=25, fontweight='bold', labelpad=20)
        ax2.set_xlim(0, final_epoch)
        ax2.set_ylim(0, 3)
        ax2.tick_params(axis='both', labelsize=20)
        ax2.axvline(x=final_epoch, color='gray', linestyle='--', linewidth=1.2, alpha=0.6)

        (loss_val_xfrac, loss_val_y), (loss_tr_xfrac, loss_tr_y) = loss_text_positions.get(
            model_name, [(0.60, 0.80), (0.60, 0.60)]
        )
        ax2.text(loss_val_xfrac * final_epoch, loss_val_y, f'validation ({final_val_loss:.4f})',
                 color='crimson', fontsize=20, fontweight='bold')
        ax2.text(loss_tr_xfrac * final_epoch, loss_tr_y, f'training ({final_train_loss:.4f})',
                 color='navy', fontsize=20, fontweight='bold')

        if model_name == "Attention-Weighted Fusion Model":
            xmax = ax2.get_xlim()[1]
            ymax = ax2.get_ylim()[1]
            ax2.text(0.25 * xmax, 1.5 * ymax, 
                     f'final model 965 epochs\nearly stopping triggered',
                     color='grey', fontsize=20, fontweight='bold', ha='left')

        ax2.spines['top'].set_visible(False)
        ax2.spines['right'].set_visible(False)

    # ===  ===
    fig.suptitle("Training and Validation Accuracy/Loss over Epochs",
                 fontsize=35, fontweight='bold', y=1)
    if save_path:
        plt.savefig(save_path, format="pdf", bbox_inches="tight")
        print(f"Figure saved to {save_path}")

    plt.show()

histories = [
    "trained_model_baseline/training_history_baseline.pkl",
    "trained_model_attention/training_history_attention.pkl",
    "trained_model_gatefusion/training_history_gatefusion.pkl",
    "trained_model_hadamard/training_history_hadamard.pkl"
]

final_epochs = [1000, 965, 1000, 1000]
model_names = [
    "Baseline Model with Late-Average Fusion",
    "Attention-Weighted Fusion Model",
    "Gate-Weighted Additive Fusion Model",
    "Hadamard Bilinear Fusion Model"
]

plot_all_histories(histories, final_epochs, model_names, save_path="training_history_plots.pdf")
```
